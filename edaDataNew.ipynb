```python
def iqr_technique(DFcolumns):
    Q1=np.percentile(DFcolumns,25)
    Q3=np.percentile(DFcolumns,75)
    IQR=Q3-Q1
    lowerrange=Q1-(1.5*IQR)
    highrange=Q3+(1.5*IQR)
    return lowerrange,highrange
```

```python
edaN.head(2)
```

```python
lower_bond,upper_bond=iqr_technique(edaN['MSRP'])
edaN=edaN[(edaN.MSRP>lower_bond) & (edaN.MSRP<upper_bond)]
```

**(7) Average minimum MSRP in each car maker brand using Scatter and Bar chart**

```python
minimum_avg_MSRP=edaN.groupby('CarMakerBrand')['MSRP'].mean().reset_index()
minimum_avg_MSRP=minimum_avg_MSRP.rename(columns={'MSRP':'avg_MSRP'})
top_10_min_avg_MSRP=minimum_avg_MSRP.nsmallest(10,'avg_MSRP')
brand_group=edaN['CarMakerBrand'].drop_duplicates()
brand_avg_sorted_with_group=pd.merge(top_10_min_avg_MSRP,brand_group,how='left')
brand_avg_sorted_with_group
```

```python
brand=brand_avg_sorted_with_group['CarMakerBrand']
MSRP=brand_avg_sorted_with_group['avg_MSRP']

plt.figure(figsize=(15,5))
plt.bar(brand,MSRP,width=0.5, color = 'orchid')
plt.xlabel('Car Maker Brand')
plt.ylabel('Average MCRP')
plt.title('Average MCRP  By CarMakerBrand')

# Show the plot
plt.show()
```
