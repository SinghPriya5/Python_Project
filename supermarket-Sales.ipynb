# Data Visualization

**1.)Distribution of profit_income in range using Histrogram**

```python
plt.figure(figsize=(10,6))
sns.set_theme(style='darkgrid')
sns.distplot(x=sales['profit_income'],color='r')
plt.xlabel("Profit Income",fontsize=12)
plt.ylabel("Density",fontsize=12)
plt.title("Distribution of profit_income on supermarket_sales",fontsize=15)
plt.show()
```

**Observations**
*1.)The range of profit income being on supermarket_sales appears to be from 1 to 49 dollars ,with majority of profit income in the range 3 to 24.*
*2.)The distribution of profit to have a peak in the 1 to 49 dollars range,with a relatively higher and lower density of sales in higher and lower profit in range.*  

**2.)Total counts in each City by using count plot.**

```python
TotalC=sales['City'].value_counts().reset_index(name='Total')
TotalC
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=sales['City'])
plt.xlabel("Total Counts",fontsize=12)
plt.ylabel("City",fontsize=12)
plt.title("Total counts in each City  on supermarket_sales",fontsize=15)
plt.show()
```

**Observations**
1.)Yangon city has the highest number of sales on supermarket_sales with over 340 sales each.
2.)Naypyitaw and Mandalay have significantly similar ,with 332 and 328 sales each.
3.)This could suggest that the demand for sales in each product for  city is not much effected. 


**3.)Average profit in each branch on supermarket_sales using point plot.**

```python
avg_proft=sales.groupby('Branch').agg({'profit_income':'mean'}).reset_index()
avg_proft.columns=['Branch','Average Profit']
avg_proft=avg_proft.round(2)
avg_proft
```

```python
from statistics import mean
plt.figure(figsize=(6,8))
sns.pointplot(x='Branch',y='profit_income',data=sales,estimator='mean')
plt.xlabel('Branch',fontsize=10)
plt.ylabel("Profit",fontsize=10)
plt.title("Average profit in each branch on supermarket_sales",fontsize=12)
plt.show()
```

**Observations**
1.)The average profit on sales in supermarket varies  varies significantely across different with A,B and C branch.
2.)Average profit increases to move the outer sales (A,B,C) from towards the center of the branch.
